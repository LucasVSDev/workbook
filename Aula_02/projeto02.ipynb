## Passo a passo do problema

- Buscar as informações da ação automaticamente
- Criar as análises solicitadas
  - Cotação máxima
  - Cotação mínima
  - Valor médio
- Enviar um e-mail automaticamente para o gestor

#### Buscar dados da ação

Bibliotecas: `yfinence` e `matplotlib`

```python
# importa biblioteca yfinace
import yfinance
```

```python
ticker = input("Digite o codigo da ação desejada: ")

dados = yfinance.Ticker(ticker).history(start="2024-01-01", end="2024-07-31")
fechamento = dados.Close
fechamento.plot()
```

### Criar as análises solicitadas
- Cotação máxima
- Cotação mínima
- Valor médio 

```python
maximo = round(fechamento.max(), 2)
minimo = round(fechamento.min(), 2)
valor_medio = round(fechamento.mean(), 2) 

print(maximo)
print(minimo)
print(valor_medio)
```

#### Enviar o e-mail automaticamente

- Abrir o navegador e ir para o Gmail
- Clicar no botão Escrever
- Digitar o email do destinatário e tecla TAB
- Digitar o assunto e teclar TAB
- Digitar a messagem
- Clicar no botão Enviar

Biblioteca: `pyautogui`, `pyperclip`, `webbrowser`

`pip install pyautogui`

#### Bibliotecas

```python
import pyautogui
import pyperclip
import webbrowser
import time

destinatario = "viniciusramos9807@gmail.com"
assunto = "Análises de Projeto 2024"

messagem = f""" 
Prezado Gestor,

segue as análises solicitadas da ação {ticker}:

Cotação Máxima: R${maximo}
Cotação Mínima: R${minimo}
Cotação Médio: R${valor_medio}

Qualquer dúvida, estou à disposição!

Atte.
"""
# Abrir o navegador e ir para o Gmail
webbrowser.open("www.gmail.com")
time.sleep(4)

# configura uma pause de 3 segundo
pyautogui.PAUSE = 3

# Clicar no botão escrever
pyautogui.click(x = 81, y = 162)

# digita o email do destinataro e tecla TAB
pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# digita o assunto do email
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")

# digita a messagem 
pyperclip.copy(messagem)
pyautogui.hotkey("ctrl", "v")

# Enviar a messagem
pyautogui.click(x=792, y=701)
pyautogui.hotkey("enter")
```
